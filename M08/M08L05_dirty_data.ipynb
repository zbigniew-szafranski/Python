# Jak wyczyścić brudne dane

```python
import pandas as pd

df = pd.read_csv(
    'data/forex_usd_data.csv',
    parse_dates=['date(y-m-d)'],
    index_col='date(y-m-d)',
)

# Dobrze jest przyjrzeć się naszym danym, czy na pewno wszystko jest w porządku?
# 1. W szczególności przyglądnijmy się kilku pierwszym i ostatnim wierszom danych.
# 2. Czy pojawiają się gdzieś wartości NaN (Not a Number - oznaczający najczęściej brak danych)?

df
```

```python
# Dobrze jest też sprawdzić jak Pandas rozumie nasze dane.
# 1. Czy na pewno liczby traktuje jako liczby a daty jako daty?

df.info()
```

```python
# Dobrze jest też przyjrzeć się podstawowym wielkościom statystycznym o naszych danych - służy do tego `.describe()`:
# 1. W szczególności czy mamy jakieś NaN (sprawdź count)?
# 2. Czy mamy jeszcze gdzieś brakujące dane zapisane jako -99, -999 albo inna specjalna wartość (sprawdź min i max)?

df.describe().T
```
