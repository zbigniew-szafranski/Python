# Obliczenia w Pandasie

```python
import pandas as pd

df = pd.read_csv(
    'data/forex_usd_data.csv',
    parse_dates=['date(y-m-d)'],
    index_col='date(y-m-d)',
).rename(columns={
    'Polish Zloty': 'PLN',
    'Euro': 'EUR',
})

# W Pandasie możesz wyciągnąć pojedynczą kolumnę. Otrzymujesz wówczas obiekt typu SERIES.

c = df['PLN']
print(type(c))
c
```

```python
# Możesz także wykonywać obliczenia między dwoma kolumnami. Np. możesz podzielić dwie kolumny przez siebie:

pln_to_eur = df['PLN'] / df['EUR']
pln_to_eur.plot(ylabel='1 EUR = X PLN');
```

```python
# W jaki sposób dokleić tak utworzone dane do naszego DataFrame? Użyj assign:

new_df = df.assign(
    PLN_EUR=df['PLN'] / df['EUR'],
)

new_df
```
