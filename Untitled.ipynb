```python
import pandas as pd
import matplotlib.pyplot as plt

def remove_thousands_separator(data: str) -> float:
    return float(data.replace(',', ''))

df = pd.read_csv('/M08/data_M08/macroeconomicsPL.csv',
                 parse_dates=['DATE'],
                 index_col='DATE',
                 dayfirst=True,
                 engine='python',
                 skipfooter=3,
                 # thousands=',',
                 na_values={
                     'SPD': '-99'},
                 ).rename(columns={
    'sr_NBP':'SR',
    'WIBOR 3M':'WIBOR'}
).assign(
    SR_CPI=lambda x: x['SR'] - x['CPI'],
    RB=lambda x: x['RB'].map(remove_thousands_separator),
    CPI_min= lambda x: x['CPI'].rolling(60).min(),
    CPI_max = lambda x: x['CPI'].rolling(60).max(),
)

df.plot(
    y=['CPI','CPI_min', 'CPI_max'],
    figsize=(15,5),
    grid=True,
    style={
        'CPI_min': 'k',
        'CPI_max': 'k',
    }
);
```
